```python
import pandas as pd
import matplotlib.pyplot as plt
import requests


#list of cryptocurrency pairs

url = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=6938&page=1&sparkline=false'
data = requests.get(url).json()
df = pd.DataFrame.from_dict(data)
df.drop(columns=["image", 
                 "market_cap_rank", 
                 "fully_diluted_valuation", 
                 "high_24h", 
                 "low_24h", 
                 "price_change_24h", 
                 "price_change_percentage_24h", 
                 "market_cap_change_24h", 
                 "market_cap_change_percentage_24h", 
                 "circulating_supply", 
                 "total_supply", 
                 "max_supply", 
                 "ath", 
                 "roi",
                 "ath_change_percentage",
                 "ath_date", 
                 "last_updated", 
                 "atl", 
                 "atl_change_percentage", 
                 "atl_date"], 
        inplace=True)
df = df.head(10)

SumData = df["market_cap"].tolist()
Labels = df["name"].tolist()

plt.pie(SumData,
        labels=Labels, 
        autopct='%1.1f%%') #Apresenta as percentagens das fatias

plt.axis('equal')
plt.title("MarketShare between Top 10 Cryptocurrencies")
#plt.legend(loc="lower right")
plt.show()


df
```
